`Dataset`: [Pakistan House Price Dataset](https://www.kaggle.com/datasets/jillanisofttech/pakistan-house-price-dataset/data)  


# **About Dataset**
## Context
The dataset is scraped from zameen.com which is Pakistan's top real estate platform. It contains listings of properties from five major cities of Pakistan.  
The aim of this project is to perform Exploratory Data Analysis (EDA) to uncover insights and use Machine Learning models to predict property price based on the given attributes.  

### **Content**
#### Column Descriptions:
`property_id`: Unique identifier for each property.  
`location_id`: Unique identifier for each location within a city.  
`page_url`: The URL of the webpage where the property was published.  
`property_type`: Categorization of the property into six types: House, FarmHouse, Upper Portion, Lower Portion, Flat, or Room.  
`price`: The price of the property, which is the dependent feature in this dataset.  
`city`: The city where the property is located. The dataset includes five cities: Lahore, Karachi, Faisalabad, Rawalpindi, and Islamabad.  
`province`: The state or province where the city is located.  
`location`: Different types of locations within each city.  
`latitude` and `longitude`: Geographic coordinates of the cities.

```python
# checking for outliers
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print(numeric_cols)
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()# auto set
plt.show()
```

## Explanation of Above box plot

1 Boxplot of Price

The median (middle line) is very low compared to the maximum, which suggests a skewed distribution.

There are many outliers on the high end (points above the whisker)  these are very expensive properties compared to the majority.

Interpretation: Most properties are in a normal price range, but there are some extremely expensive properties that may affect the model if not handled.

2 Boxplot of Latitude

Most data is clustered around a narrow latitude range (e.g., 2435), which corresponds to the geographic locations of the cities in Pakistan.

A few outliers above or below this range  could be misentered coordinates.

3 Boxplot of Longitude

Most properties are around 6675 longitude.

Outliers (very low or high longitude) could indicate incorrect data entries or far-away locations that are unusual for the dataset.

4 Boxplot of Baths

Median bathrooms are around 24.

Outliers exist with extremely high values (e.g., 20+ bathrooms or even 400 in your data).

These extreme outliers are likely data errors and should probably be handled before modeling.

5 Boxplot of Bedrooms

Most properties have 25 bedrooms (median ~3).

There are a few properties with very high bedroom counts (1070+)  again, likely data errors or extremely large properties.

6 Boxplot of Area Size

Most property areas are within a certain range (smallmedium plots).

Some extreme outliers are very large (e.g., 700+ in your units), which could skew the model if not addressed.

```python
# Identifying outliers using IQR method
for col in numeric_cols:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f'{col}: {len(outliers)} outliers')
```

##### Point to note:  
The dataset contains outliers

```python
# removing outliers
for col in ['Area Size', 'price', 'baths', 'bedrooms']:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

### **Feature Engineering**

```python
#creating new feature for area size in square feet
df['area_sqft'] = np.where(
    df['Area Type'] == 'Marla', df['Area Size'] * 272,
    np.where(df['Area Type'] == 'Kanal', df['Area Size'] * 5445, df['Area Size'])
)
```

```python
df['price_per_sqft'] = df['price'] / df['area_sqft']
```

```python
# adding column for year
df['year_added'] = df['date_added'].dt.year
```
